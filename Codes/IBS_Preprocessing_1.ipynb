```python
count=1
da=[]
for i in Id:
              # Downloading...
        print(count)
        net_handle = Entrez.efetch(
            db="nucleotide", id=i, rettype="fasta", retmode="text"
        )
        ref = SeqIO.read(  net_handle,"fasta")
        da.append([i,ref.seq])
        count+=1
out_handle = open("./Dataset/"+filename, "w")
writer = csv.writer(out_handle,lineterminator='\n',)
writer.writerows(da)
out_handle.close()
net_handle.close()
print("Saved")
```
